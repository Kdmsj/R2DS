```python
file_path = "transactions.txt"

with open(file_path, "r") as file:
    transactions = [int(line.strip()) for line in file]

transactions
```

```python
total_sum = sum(transactions)
total_sum
```

```python
num_transactions = len(transactions)
num_transactions
```

```python
average_transaction = total_sum / num_transactions if num_transactions > 0 else 0
average_transaction
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(range(len(transactions)), transactions, color='red')
plt.xlabel("Transaction Index")
plt.ylabel("Amount")
plt.title("Transaction Values Over Time")
plt.show()
```
